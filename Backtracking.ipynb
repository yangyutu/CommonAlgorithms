## Basic level of backtracking

### General template

```python
The honest working code is given below. We include a global finished flag to allow for premature termination, which could be set in any application-specific routine.
bool finished = FALSE;               /* found all solutions yet? */

backtrack(int a[], int k, data input)
{
        int c[MAXCANDIDATES];        /* candidates for next position */
        int ncandidates;             /* next position candidate count */
        int i;                       /* counter */

        if (is_a_solution(a,k,input))
                process_solution(a,k,input);
        else {
                k = k+1;
                construct_candidates(a,k,input,c,&ncandidates);
                for (i=0; i<ncandidates; i++) {
                        a[k] = c[i];
                        backtrack(a,k,input);
                        if (finished) return;   /* terminate early */
                }
        }
}
```

### Subset

```python
# 78. Subsets
# https://leetcode.com/problems/subsets/
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        
        def help(i, temp):
            res.append(list(temp)) # step1. process solution Store the result at every node.
            if i == len(nums): # step 2 termination condition
                return
            for j in range(i, len(nums)): # step 3 construct more candidates
                temp.append(nums[j])
                help(j + 1, temp)
                temp.pop()     
        res = []    
        help(0, [])
        return res

class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        
        def help(i, temp):  
            if i == len(nums):
                res.append(list(temp)) # only store the result at the end
                return
            if i > len(nums):
                return
            temp.append(nums[i])
            help(i + 1, temp)
            temp.pop()
            help(i + 1, temp)
        res = []
        help(0, [])
        return res
```

### 77. Combinations

```python
77. Combinations
# Given two integers n and k, return all possible combinations of k numbers out of 1 ... n.

class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:      
        def help(temp, i):
            # step 1 process solution
            if len(temp) == k:
                res.append(list(temp))
                return
            # step 2 early termination condition
            if i > n:
                return
            
            # step 3 construct more solutions
            # one way to grow a tree, select and skip the i element
            temp.append(i) # since we are 1 based
            help(temp, i + 1)
            temp.pop()
            help(temp, i + 1)
        res = []
        help([], 1)
        return res
    
    
class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
        def help(temp, i):
            if len(temp) == k:
                res.append(list(temp))
                return
            
            if i == n + 1:
                return
            
            # anther way to grow the tree, skip different elements    
            for j in range(i, n + 1):
                temp.append(j) # since we are 1 based
                help(temp, j + 1)
                temp.pop()
        res = []
        help([], 1)
        return res    
```

### 46. Permutations

```python

#46. Permutations
#https://leetcode.com/problems/permutations/

# a naive solution
class Solution:
    """
    @param: nums: A list of integers.
    @return: A list of permutations.
    """
    def permute(self, nums):
        # write your code here
        
        # level can be removed since len(temp) == level
        def help(level, temp):
            
            # step1: process solution
            if len(temp) == len(nums):
                if len(set(temp)) == len(nums):
                    res.append(temp[:])
                return    
            # step2: early termination
            if level == len(nums):
                return
            # step3: construct other candidates
            for j in range(len(nums)):
                
                temp.append(nums[j])
                help(level + 1, temp)
                temp.pop()
                
        res = []
        help(0, [])
        return res





def permute(self, nums):
    def backtrack(start, end):
        if start == end:
            ans.append(nums[:])
        for i in range(start, end):
            nums[start], nums[i] = nums[i], nums[start]
            backtrack(start+1, end)
            nums[start], nums[i] = nums[i], nums[start]

    ans = []
    backtrack(0, len(nums))
    return ans   


class Solution:
    def permute(self, nums: List[int]) -> List[List[int]]:
      
        def help(temp, used):
            if len(temp) == len(nums):
                res.append(list(temp))
                return               
            for j in range(len(nums)):
                if used[j] == 0:
                    temp.append(nums[j])
                    used[j] = 1
                    help(temp, used)
                    used[j] = 0
                    temp.pop()
                    
        used = [0 for _ in range(len(nums))]
        res = []
        help([], used)
        
        return res
```

```python
6. Permutations II
Given a list of numbers with duplicate number in it. Find all unique permutations.

# A dictionary solution
class Solution:
    """
    @param: :  A list of integers
    @return: A list of unique permutations
    """

    def permuteUnique(self, nums):
        # write your code here
        
        def help(count, temp):
            
            if len(temp) == len(nums):
                res.append(temp[:])
            else:
            
                for x in count: # here we must iterate over dict keys instead of nums
                    if count[x] > 0:
                        count[x] -= 1 
                        temp.append(x)
                        help(count, temp)
                        temp.pop()
                        count[x] += 1
                    
        
        nums.sort()
        res = []
        count = collections.Counter(nums)
        help(count, [])
        
        return res

class Solution:
    """
    @param: :  A list of integers
    @return: A list of unique permutations
    """

    def permuteUnique(self, nums):
        # write your code here
        
        def help(visited, temp):
            
            if len(temp) == len(nums):
                res.append(temp[:])
            else:
            
                for i in range(len(nums)):
                    if visited[i]:
                        continue
                    # a' a" b
                    # => a' a" b => √
                    # => a" a' b => x
                    # 不能跳过一个a选下一个a
                    if i > 0 and nums[i] == nums[i - 1] and not visited[i - 1]:
                        continue
                    
                    visited[i] = True
                    temp.append(nums[i])
                    help(visited, temp)
                    temp.pop()
                    visited[i] = False
                    
        
        nums.sort()
        
        res = []
        visited = [False] * len(nums)
        help(visited, [])
        
        return res
```

### 526. Beautiful Arrangement

Suppose you have n integers labeled 1 through n. A permutation of those n integers perm (1-indexed) is considered a beautiful arrangement if for every i (1 <= i <= n), either of the following is true:

Given an integer n, return the number of the beautiful arrangements that you can construct.

```python
class Solution:
    def countArrangement(self, n: int) -> int:
        
        
        visited = [False] * (n + 1)
        
        def helper(i, visited):
            
            if i == n + 1:
                self.count += 1
            else:
                for j in range(1, n + 1):
                    if not visited[j]:
                        if j % i == 0 or i % j == 0:
                            visited[j] = True
                            helper(i + 1, visited)
                            visited[j] = False
        self.count = 0
        helper(1, visited)
        return self.count
```

Given the root of a binary tree and an integer targetSum, return all root-to-leaf paths where the sum of the node values in the path equals targetSum. Each path should be returned as a list of the node values, not node references.

A root-to-leaf path is a path starting from the root and ending at any leaf node. A leaf is a node with no children.

```python
# element to consider when choosing different patterns
# whehter graph root is fixed or can be chosen
# whether we would like add value first, or defer to later

# pattern one
class Solution:
    def pathSum(self, root: Optional[TreeNode], targetSum: int) -> List[List[int]]:
        
        if not root:
            return []
        
        def helper(temp_sum, temp, node):
            
            if not node.left and not node.right:
                if temp_sum + node.val == targetSum:
                    res.append(temp[:] + [node.val])
            else:
                # pattern one
                temp.append(node.val)
                for child in node.left, node.right:
                    if child:
                        helper(temp_sum + node.val, temp, child)
                temp.pop()
        
        
        res = []
        # pattern 1
        helper(0, [], root)
        return res
    
# pattern two
class Solution:
    def pathSum(self, root: Optional[TreeNode], targetSum: int) -> List[List[int]]:
        
        if not root:
            return []
        
        def helper(temp_sum, temp, node):
            
            if not node.left and not node.right:
                if temp_sum == targetSum:
                    res.append(temp[:])
            else:
                
                for child in node.left, node.right:
                    if child:
                        # pattern one
                        temp.append(child.val)
                        helper(temp_sum + child.val, temp, child)
                        temp.pop()
        
        
        res = []
        # pattern 2
        helper(root.val, [root.val], root)
        return res
```

### 698. Partition to K Equal Sum Subsets

```python
Given an array of integers nums and a positive integer k, find whether it's possible to divide this array into k non-empty subsets whose sums are all equal.

 

Example 1:

Input: nums = [4, 3, 2, 3, 5, 2, 1], k = 4
Output: True
Explanation: It's possible to divide it into 4 subsets (5), (1, 4), (2,3), (2,3) with equal sums.
    
class Solution:
    def canPartitionKSubsets(self, nums: List[int], k: int) -> bool:
        totalSum = sum(nums) 
        subSumTarget = totalSum // k
        if k * subSumTarget != totalSum:
            return False
        
        
        def help(subSumTarget, sumGroup, k, idx):
            
            if len(nums) == idx:
                return True
            else:
                
                for i in range(k):
                    if sumGroup[i] + nums[idx] <= subSumTarget:
                        sumGroup[i] += nums[idx]
                        if help(subSumTarget, sumGroup, k, idx + 1):
                            return True
                        sumGroup[i] -= nums[idx]
                        if sumGroup[i] == 0: # because order does not matter, only need to search once for each element
                            break
                return False
        # additional steps to reduce complexity
        nums.sort()
        if nums[-1] > subSumTarget: return False
        while nums and nums[-1] == subSumTarget:
            nums.pop()
            k -= 1
        nums.reverse() # consider large elements first reduce the branching at the beginning
        return help(subSumTarget, [0] * k, k, 0)    
```

### 679. 24 Game
You have 4 cards each containing a number from 1 to 9. You need to judge whether they could operated through *, /, +, -, (, ) to get the value of 24.

```python
class Solution:
    def judgePoint24(self, operands: List[int]) -> bool:
                
        if len(operands) == 1:
            if abs(operands[0] - 24) < 1e-6:
                return True
            else:
                return False
        else:
            operandsCopy = [e for e in operands]
            for i in range(len(operands)):
                for j in range(len(operands)):
                    if i != j:
                        operandsCopy.remove(operands[i])
                        operandsCopy.remove(operands[j])
                        for operator in ["+", "-", "*", "."]:
                            if operator == '.' and operands[j] == 0:
                                continue
                            if operator == '+':
                                temp = operands[i] + operands[j]
                            elif operator == '-':
                                temp = operands[i] - operands[j]
                            elif operator == '*':
                                temp = operands[i] * operands[j]   
                            elif operator == '.':
                                temp = operands[i] / operands[j]
                            operandsCopy.append(temp)
                            if self.judgePoint24(operandsCopy):
                                return True
                            operandsCopy.pop()
                        operandsCopy.append(operands[i])
                        operandsCopy.append(operands[j])
                            
            return False
```

```python
class Solution:
    def uniquePathsIII(self, grid: List[List[int]]) -> int:
        
        
        def back_track(i, j, visited, remaining):
            
            if (i, j) == target and remaining == 0:
                self.ans += 1
            else:
                
                visited.add((i, j))
                
                for dx, dy in (-1, 0), (1, 0), (0, -1), (0, 1):
                    new_x = i + dx
                    new_y = j + dy
                    if 0 <= new_x < len(grid) and 0 <= new_y < len(grid[0]):
                        if (new_x, new_y) not in visited and grid[new_x][new_y] != -1:
                            back_track(new_x, new_y, visited, remaining - 1)
                visited.discard((i, j))
                
        remaining = len(grid) * len(grid[0])
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == 2:
                    target = (i, j)
                elif grid[i][j] == -1:
                    remaining -= 1
                elif grid[i][j] == 1:
                    start = (i, j)
                    
        self.ans = 0
        back_track(start[0], start[1], set(), remaining -1)
        
        return self.ans
```

### 494. Target Sum
You are given a list of non-negative integers, a1, a2, ..., an, and a target, S. Now you have 2 symbols + and -. For each integer, you should choose one from + and - as its new symbol.

Find out how many ways to assign symbols to make sum of integers equal to target S.

```python
class Solution:
    def findTargetSumWays(self, nums: List[int], S: int) -> int:
        
        
        memo = {}
        
        def help(i, target):
            
            if i == len(nums):
                if target == 0:
                    return 1
                else:
                    return 0
            else:
                if (i, target) in memo:
                    return memo[(i, target)]
                
                ways = help(i + 1, target + nums[i]) + help(i + 1, target - nums[i])
                
                memo[(i, target)] = ways
                
                return ways
                
                
        return help(0, S)
                
```

If the frog's last jump was k units, then its next jump must be either k - 1, k, or k + 1 units. Note that the frog can only jump in the forward direction.

```python
class Solution:
    def canCross(self, stones: List[int]) -> bool:
        
        memo = {}
        def dfs(i, k):
            if (i, k) in memo:
                return memo[i, k]
            if i == len(stones) - 1:
                return True
            else:
                for j in range(i + 1, len(stones)):
                    if stones[j] - stones[i] in (k - 1, k, k + 1):
                        if dfs(j, stones[j] - stones[i]):
                            memo[j, stones[j] - stones[i]] = True
                            return True
                    if stones[j] - stones[i] > k + 1:
                        break
                memo[i, k] = False
                return False

        if stones[1] - stones[0] != 1:
            return False
        
        return dfs(1, 1)

class Solution:
    def canCross(self, stones: List[int]) -> bool:
        
        self.memo ={}
        return self.dfs(stones[0], 0, stones, [1])
    
    
    def dfs(self, pos, idx, stones, jumps):
        
        if idx == len(stones) - 1:
            return True
        else:
            
            state = tuple([idx] + jumps)
            if state in self.memo:
                return self.memo[state]
            
            # the start idx can be further optimized by binary search
            for j in range(idx + 1, len(stones)):
                
                jumpSize = stones[j] - pos
                if jumpSize > max(jumps):
                    break
                if jumpSize in jumps:
                    newJumps = [jumpSize - 1, jumpSize, jumpSize + 1]
                    if self.dfs(stones[j], j, stones, newJumps):
                        self.memo[state] = True
                        return True
            self.memo[state] = False
            return False
        
```

### 10. Regular Expression Matching

Given an input string s and a pattern p, implement regular expression matching with support for '.' and '*' where:

The matching should cover the entire input string (not partial).

### 464. Can I Win

In the "100 game" two players take turns adding, to a running total, any integer from 1 to 10. The player who first causes the running total to reach or exceed 100 wins.

What if we change the game so that players cannot re-use integers?

For example, two players might take turns drawing from a common pool of numbers from 1 to 15 without replacement until they reach a total >= 100.

Given two integers maxChoosableInteger and desiredTotal, return true if the first player to move can force a win, otherwise, return false. Assume both players play optimally.

```python
class Solution:
    def canIWin(self, maxInt: int, total: int) -> bool:
        
        if maxInt > total:
            return True
        # Total is too large, nobody can win
        if (maxInt + 1) * maxInt // 2 < total:
            return False

        remain = [e for e in range(1, maxInt + 1)]
        
        return self.dfs(0, total, remain, {})
        
        
    def dfs(self, currSum, total, remain, memo):
        
        for e in remain:
            if currSum + e >= total:
                return True
            
        if tuple(remain) in memo:
            return memo[tuple(remain)]
        
        for e in remain:
            newList = list(remain)
            newList.remove(e)
            # if opponent cannot win, I win
            if not self.dfs(currSum + e, total, newList, memo):
                memo[tuple(remain)] = True
                return True
        # if there is one chance opponent can win
        # I lose
        memo[tuple(remain)] = False
        return False
```

### 291. Word Pattern II
Given a pattern and a string s, return true if s matches the pattern.

A string s matches a pattern if there is some bijective mapping of single characters to strings such that if each character in pattern is replaced by the string it maps to, then the resulting string is s. A bijective mapping means that no two characters map to the same string, and no character maps to two different strings.

```python
class Solution:
    def wordPatternMatch(self, p: str, s: str) -> bool:
        
        
        def helper(i, j, mappings, reverse_set):
            
            if i == len(p) and j == len(s):
                return True
            elif i < len(p) and j < len(s):
                if p[i] in mappings:
                    word = mappings[p[i]]
                    if s[j: j + len(word)] == word and helper(i + 1, j + len(word), mappings, reverse_set):
                        return True
                else:
                    for k in range(j + 1, len(s) + 1):
                        word = s[j: k]
                        if word not in reverse_set:
                            mappings[p[i]] = word
                            reverse_set.add(word)
                            if helper(i + 1, k, mappings, reverse_set):
                                return True
                            del mappings[p[i]]
                            reverse_set.discard(word)
                    
                    
                    return False
            else:
                return False
            
        return helper(0, 0, {}, set())
```

### 211. Design Add and Search Words Data Structure (using stack for iterative DFS for path existence)

Design a data structure that supports adding new words and finding if a string matches any previously added string.

Implement the WordDictionary class:

WordDictionary() Initializes the object.
void addWord(word) Adds word to the data structure, it can be matched later.
bool search(word) Returns true if there is any string in the data structure that matches word or false otherwise. word may contain dots '.' where dots can be matched with any letter.

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_word = False

class WordDictionary:

    def __init__(self):
        
        self.root = TrieNode()

    def addWord(self, word: str) -> None:
        
        curr = self.root
        for c in word:
            if c not in curr.children:
                curr.children[c] = TrieNode()
            curr = curr.children[c]
        curr.is_word = True
        
    def search(self, word: str) -> bool:
        
        stack = [(0, self.root)]
        
        while stack:
            idx, node = stack.pop()
            if idx == len(word):
                if node.is_word:
                    return True
                else:
                    continue
                # when we use stack iterative dfs, we cannot return False in the while loop
                #else:
                #    return False
            
            if word[idx] == '.':
                for c in node.children:
                    stack.append((idx + 1, node.children[c]))
            else:
                if word[idx] in node.children:
                    stack.append((idx + 1, node.children[word[idx]]))
                    
        return False
```

### 1367. Linked List in Binary Tree

Given a binary tree root and a linked list with head as the first node. 

Return True if all the elements in the linked list starting from the head correspond to some downward path connected in the binary tree otherwise return False.

In this context downward path means a path that starts at some node and goes downwards.

```python
# back_track inside dfs
class Solution:
    def isSubPath(self, head: Optional[ListNode], root: Optional[TreeNode]) -> bool:
        
        def back_track(curr, start):
            
            if not curr:
                return True
            else:
                if not start or start.val != curr.val:
                    return False
            
                if back_track(curr.next, start.left) or back_track(curr.next, start.right):
                    return True
                return False        
        
        def find_head(head, root):
            if not root:
                return False
            if head.val == root.val:
                if back_track(head, root):
                    return True
            
            if find_head(head, root.left) or find_head(head, root.right):
                return True

            return False
        
        return find_head(head, root)
```

### 780. Reaching Points

Given four integers sx, sy, tx, and ty, return true if it is possible to convert the point (sx, sy) to the point (tx, ty) through some operations, or false otherwise.

The allowed operation on some point (x, y) is to convert it to either (x, x + y) or (x + y, y).

```python
class Solution:
    def reachingPoints(self, sx: int, sy: int, tx: int, ty: int) -> bool:
        
        while tx > sx and ty > sy:
            if tx > ty:
                tx = tx % ty # this is equivalent to move up in a single direction until cannot
            else:
                ty = ty % tx
        
        # now we are hitting the leftmost branch or the rightmost branch of the binary tree
        if tx == sx and ty >= sy and (ty - sy) % sx == 0:
            return True
        return ty == sy and tx >= sx and (tx - sx) % sy == 0
```

### 698. Partition to K Equal Sum Subsets

Given an integer array nums and an integer k, return true if it is possible to divide this array into k non-empty subsets whose sums are all equal.

```python
class Solution:
    def canPartitionKSubsets(self, nums: List[int], k: int) -> bool:
        
        total_sum = sum(nums)
        subsum = total_sum // k
        if total_sum % k != 0:
            return False
        
        
        def back_track(buckets, i):
            
            if i == len(nums):
                return True
            else:
               
                for bucket_id in range(k):
                    if buckets[bucket_id] + nums[i] <= subsum:
                        buckets[bucket_id] += nums[i]

                        if back_track(buckets, i + 1):
                            return True

                        buckets[bucket_id] -= nums[i]
        
                return False
        
        buckets = [0] * k
        nums.sort(reverse=True)
        if back_track(buckets, 0):
            return True
        else:
            return False
```
