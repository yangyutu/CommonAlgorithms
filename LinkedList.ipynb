### 82. Remove Duplicates from Sorted List II
Given a sorted linked list, delete all nodes that have duplicate numbers, leaving only distinct numbers from the original list.

For example,
Given 1->2->3->3->4->4->5, return 1->2->5.
Given 1->1->1->2->3, return 2->3.

```python
class Solution:
    def deleteDuplicates(self, head: ListNode) -> ListNode:
        
        dummy = ListNode(-1)
        dummy.next = head
        prev = dummy
        
        while head and head.next:
            if head.val == head.next.val:
                while head.next and head.val == head.next.val:
                    head = head.next
                prev.next = head.next
                head = head.next
            else:
                prev = head
                head = head.next
            
        return dummy.next
    
    
class Solution:
    def deleteDuplicates(self, head: ListNode) -> ListNode:
        
        tail = dummy = ListNode(-1)
        while head:    
            if not head.next or head.val != head.next.val:
                tail.next = head
                head = head.next
                tail = tail.next
                tail.next = None
            else:
                while head.next and head.val == head.next.val: # move to the end the same block
                    head = head.next        
                head = head.next
        return dummy.next    
```

### 328. Odd Even Linked List

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def oddEvenList(self, head: ListNode) -> ListNode:
        if not head:
            return head
        
        evenHead = head.next
        
        odd = head
        even = head.next
        
        while even and even.next:
            odd.next = even.next
            odd = odd.next
            even.next = odd.next
            even = even.next

        odd.next = evenHead
        
        return head
            
        
```

### 61. Rotate List

```python
class Solution:
    def rotateRight(self, head: ListNode, k: int) -> ListNode:
        
        if not head:
            return head
        curr = head
        length = 1
        
        while curr.next:
            length += 1
            curr = curr.next
        tail = curr
        k = k % length
        
        curr = head
        for _ in range(length - k - 1):
            curr = curr.next
        
        tail.next = head
        newHead = curr.next
        curr.next = None
        return newHead
```

### 708. Insert into a Sorted Circular Linked List

```python
class Solution:
    def insert(self, head: 'Node', insertVal: int) -> 'Node':
        newNode = ListNode(insertVal)
                
        if not head:
            head = newNode
            head.next = head
            return head
        # single node has been considered in the following loop
        #if head.next == head:
        #    head.next = newNode
        #    head.next.next = head
        #    return head
            
        
        curr = head
        
        prev = None
        kink = head # kink will be the largest element
        while True:
            prev = curr
            curr = curr.next
            if curr.val <= prev.val:
                kink = prev
            if prev.val <= insertVal <= curr.val:
                prev.next = newNode
                newNode.next = curr
                break
            if curr == head: # cannot find the position, insert after the kink
                newNode.next = kink.next
                kink.next = newNode
                break
        return head
    
    
        
```

### 86. Partition List
Given a linked list and a value x, partition it such that all nodes less than x come before nodes greater than or equal to x.

You should preserve the original relative order of the nodes in each of the two partitions.

```python
class Solution:
    def partition(self, head: ListNode, x: int) -> ListNode:
        
        first = dummyFirst = ListNode(-1)
        second = dummySecond = ListNode(-1)
        while head:
            temp = head.next
            if head.val < x:
                first.next = head
                first = first.next
                first.next = None # to avoid cycle
            else:
                second.next = head
                second = second.next
                second.next = None
            head = temp
        
        first.next = dummySecond.next
        return dummyFirst.next
```

### 146. LRU Cache

get(key) - Get the value (will always be positive) of the key if the key exists in the cache, otherwise return -1.
put(key, value) - Set or insert the value if the key is not already present. When the cache reached its capacity, it should invalidate the least recently used item before inserting a new item.

```python
class DListNode:
    def __init__(self, key, val):
        self.key = key
        self.val = val
        self.prev, self.next = None, None
        
class LRUCache:

    def __init__(self, capacity: int):
        
        self.capacity = capacity
        self.head, self.tail = DListNode(-1, -1), DListNode(-1, -1)
        self.head.next, self.tail.prev = self.tail, self.head
        self.keyMap = {}
        
    def _add_node(self, node):
        
        
        node.prev = self.head
        node.next = self.head.next
        
        self.head.next.prev, self.head.next = node, node
        
    def _remove_node(self, node):
        
        node.prev.next, node.next.prev = node.next, node.prev
        
    def _move_to_front(self, node):
        
        self._remove_node(node)
        self._add_node(node)
        
        
    def get(self, key: int) -> int:
        
        if key not in self.keyMap:
            return -1
        
        node = self.keyMap[key]
        
        self._move_to_front(node)
        
        return node.val
        
        
    def put(self, key: int, value: int) -> None:
        
        if key not in self.keyMap:
            node = DListNode(key, value)

            self.keyMap[key] = node
            self._add_node(node)

            if len(self.keyMap) > self.capacity:

                tailNode = self.tail.prev
                del self.keyMap[tailNode.key]
                self._remove_node(tailNode)
        else:
            self.keyMap[key].val = value
            self._move_to_front(self.keyMap[key])
            

struct DNode {
   
   DNode(int k0, int v0) {
       val = v0;
       key = k0;
   }
    
   int val, key;
   DNode* prev = nullptr;
   DNode* next = nullptr;  
};



class LRUCache {
public:
    LRUCache(int cap) {
        capacity = cap;
        head = new DNode(-1, -1);
        tail = new DNode(-1, -1);
        head->next = tail;
        tail->prev = head;
    }
    
    int get(int key) {
        if (pos.find(key) != pos.end()) {
            DNode *node = pos[key];
            moveToHead(node);
            return node->val;
        } else {
            return -1;
        }
        
    }
    
    void put(int key, int value) {
        if (pos.find(key) != pos.end()) {
            pos[key]->val = value;
            moveToHead(pos[key]);
        } else
        {
            DNode *newNode = new DNode(key, value);
            pos[key] = newNode;
            addNode(newNode);
            
            if (pos.size() > capacity) {
                DNode *finalNode = tail->prev;
                removeNode(finalNode);
                pos.erase(finalNode->key);
            }
        }
    }
    
    void addNode(DNode *node) {
        node->prev = head;
        node->next = head->next;
        head->next->prev = node;
        head->next = node;
    }
    
    void removeNode(DNode *node) {
        node->prev->next = node->next;
        node->next->prev = node->prev;
    }
    
    void moveToHead(DNode *node){
        removeNode(node);
        addNode(node);
    }
    
    DNode *head;
    DNode *tail;
    int capacity;
    int size;
    unordered_map<int, DNode*> pos;
};

/**
 * Your LRUCache object will be instantiated and called as such:
 * LRUCache* obj = new LRUCache(capacity);
 * int param_1 = obj->get(key);
 * obj->put(key,value);
 */
```
