## Two pointers in opposite directions: two sum series

### 1.2Sum (sorted array)
Given an array of integers that is already sorted in ascending order, find two numbers such that they add up to a specific target number.
The function twoSum should return indices of the two numbers such that they add up to the target,where index1 must be less than index2.

```python
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        
        left, right = 0, len(numbers) - 1
        
        while left < right: # cannot be left <= right, it would use one number twice
            
            if numbers[left] + numbers[right] < target: 
                left += 1
            elif numbers[left] + numbers[right] > target:
                right -= 1
            else:
                return [left + 1, right + 1]
            
        return [0, 0]
```

### 1099. Two Sum Largest Sum Less Than K
Given an array A of integers and integer K, return the maximum S such that there exists i < j with A[i] + A[j] = S and S < K. If no i, j exist satisfying this equation, return -1.

```python
class Solution:
    def twoSumLessThanK(self, A: List[int], K: int) -> int:
        
        A.sort()
        left, right = 0, len(A) - 1
        
        answer = -float('inf')
        while left < right:
            
            tempSum = A[left] + A[right]
            if tempSum < K:
                answer = max(answer, tempSum)
                left += 1
            else:
                right -= 1
                
        return answer if answer != -float('inf') else -1
    
                
```

### 533. Two Sum - Closest to target

```python
533. Two Sum - Closest to target
Given an array nums of n integers, find two integers in nums such that the sum is closest to a given number, target.

Return the absolute value of difference between the sum of the two integers and the target.

Input:  nums = [-1, 2, 1, -4] and target = 4
Output: 1
Explanation:
The minimum difference is 1. (4 - (2 + 1) = 1).

class Solution:
    """
    @param nums: an integer array
    @param target: An integer
    @return: the difference between the sum and the target
    """
    def twoSumClosest(self, nums, target):
        # write your code here
        nums.sort()
        left, right = 0, len(nums) - 1
        diff = float('inf')
        while left < right:
            sum = nums[left] + nums[right]
            diff = min(diff, abs(sum - target))
            if sum == target:
                return 0
            elif sum < target:
                left += 1
            elif sum > target:
                right -= 1
                
        return diff
```

You may return any answer array that satisfies this condition.

```python
class Solution:
    def sortArrayByParity(self, A: List[int]) -> List[int]:
        
        left, right = 0, len(A) - 1
        
        while left < right:
            
            if A[left] & 1 == 1 and A[right] & 1 == 0:
                A[left], A[right] = A[right], A[left]
            if A[left] & 1 == 0:
                left += 1
            if A[right] & 1 == 1:
                right -= 1
                
        return A
```

### 125. Valid Palindrome

```python
125. Valid Palindrome

Given a string, determine if it is a palindrome, considering only alphanumeric characters and ignoring cases.

Note: For the purpose of this problem, we define empty string as valid palindrome.

class Solution:
    def isPalindrome(self, s: str) -> bool:
        
        left, right = 0, len(s) - 1
        
        
        while left < right:
            
            while left < right and not s[left].isalnum():
                left += 1
            while left < right and not s[right].isalnum():
                right -= 1
            if s[left].lower() == s[right].lower():
                left, right = left + 1, right - 1
            else:
                return False
            
        return True
        
```

```python
680. Valid Palindrome II

Given a non-empty string s, you may delete at most one character. Judge whether you can make it a palindrome.

class Solution:
    def validPalindrome(self, s: str) -> bool:
        
        i, j = 0, len(s) - 1
        while i < j:
            
            if s[i] != s[j]:
                return self.isPalind(s, i + 1, j) or self.isPalind(s, i, j - 1)
            else:
                i += 1
                j -= 1
            
        return True
    
    def isPalind(self, s, i, j):
        
        while i < j and s[i] == s[j]:
            i += 1
            j -= 1
        return i >= j
```

### 11. Container With Most Water

```python
Given n non-negative integers a1, a2, ..., an , where each represents a point at coordinate (i, ai). n vertical lines are drawn such that the two endpoints of line i is at (i, ai) and (i, 0). Find two lines, which together with x-axis forms a container, such that the container contains the most water.

Note: You may not slant the container and n is at least 2.
    
class Solution:
    def maxArea(self, height: List[int]) -> int:
        
        res = 0
        
        left, right = 0, len(height) - 1
        # The intuition is double loop with early stop and start condition
        while left < right:
            
            res = max(res, (right - left) * min(height[left], height[right]))
            
            if height[left] < height[right]:# no need to move right since moving right can only reduce the result
                left += 1
            else:
                right -= 1
                
        return res
```

### 360. Sort Transformed Array

```python
Given a sorted array of integers nums and integer values a, b and c. Apply a quadratic function of the form f(x) = ax2 + bx + c to each element x in the array.

The returned array must be in sorted order.

Expected time complexity: O(n)

Example 1:

Input: nums = [-4,-2,2,4], a = 1, b = 3, c = 5
Output: [3,9,15,33]
Example 2:

Input: nums = [-4,-2,2,4], a = -1, b = 3, c = 5
Output: [-23,-5,1,7]


class Solution:
    def sortTransformedArray(self, nums: List[int], a: int, b: int, c: int) -> List[int]:
        
        nums = [a * e * e + b * e +c for e in nums]
        
        left, right = 0, len(nums) - 1
        
        res = [0] * len(nums)
        
        
        # assume a >= 0
        # if a = 0 and we have a linear equation, the following also work
        # when a >=0, we have two reverse sorted array starting from each end
        if a >=0:
            idx = right
            while left <= right:

                if nums[left] > nums[right]:
                    res[idx] = nums[left]
                    left += 1
                else:
                    res[idx] = nums[right]
                    right -= 1

                idx -= 1
        else:
            # when a <0, we have two sorted arrays starting from each end
            idx = 0
            while left <= right:

                if nums[left] < nums[right]:
                    res[idx] = nums[left]
                    left += 1
                else:
                    res[idx] = nums[right]
                    right -= 1

                idx += 1
            
        return res
```

### Lintcode 399. Nuts & Bolts Problem (advanced application of quicksort and parition)

```python
399. Nuts & Bolts Problem
English
Given a set of n nuts of different sizes and n bolts of different sizes. There is a one-one mapping between nuts and bolts.

Comparison of a nut to another nut or a bolt to another bolt is not allowed. It means nut can only be compared with bolt and bolt can only be compared with nut to see which one is bigger/smaller. We will give you a compare function to compare nut with bolt.

Using the function we give you, you are supposed to sort nuts or bolts, so that they can map in order.


"""
class Comparator:
    def cmp(self, a, b)
You can use Compare.cmp(a, b) to compare nuts "a" and bolts "b",
if "a" is bigger than "b", it will return 1, else if they are equal,
it will return 0, else if "a" is smaller than "b", it will return -1.
When "a" is not a nut or "b" is not a bolt, it will return 2, which is not valid.
"""

import random

class Solution:
    # @param nuts: a list of integers
    # @param bolts: a list of integers
    # @param compare: a instance of Comparator
    # @return: nothing
    def sortNutsAndBolts(self, nuts, bolts, compare):
        # write your code here
        
        start, end = 0, len(nuts) - 1
        
        
        def doubleQuickSort(start, end):
            
            if start >= end:
                return
            
            pivot_N = random.randint(start, end)
            
            for i in range(start, end + 1):
                if compare.cmp(nuts[pivot_N], bolts[i]) == 0 or compare.cmp(bolts[i], nuts[pivot_N]) == 0:
                    pivot_B = i
                    break
                
            pivot_N = partition(nuts, start, end, bolts[pivot_B], pivot_N)
            pivot_B = partition(bolts, start, end, nuts[pivot_N], pivot_B)
            
            doubleQuickSort(start, pivot_N - 1)
            doubleQuickSort(pivot_N + 1, end)
    
        def partition(A, left, right, cmpObject, pivotIdx):
            
            A[right], A[pivotIdx] = A[pivotIdx], A[right]
            
            
            slow = left
            
            for fast in range(left, right):
                
                if compare.cmp(A[fast], cmpObject) == -1 or compare.cmp(cmpObject, A[fast]) == 1:
                    A[fast], A[slow] = A[slow], A[fast]
                    slow += 1
            
            A[slow], A[right] = A[right], A[slow]
            return slow
            
        doubleQuickSort(start, end)
```

```python
215. Kth Largest Element in an Array

Find the kth largest element in an unsorted array. Note that it is the kth largest element in the sorted order, not the kth distinct element.

class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        
        def help(start, end, k):
            
            if start == end:
                return nums[start] 
            
            idx = random.randint(start, end)
            
            nums[end], nums[idx] = nums[idx], nums[end]
            
            slow = start
            
            for i in range(start, end):
                if nums[i] < nums[end]:
                    nums[slow], nums[i] = nums[i], nums[slow]
                    slow += 1
            
            nums[slow], nums[end] = nums[end], nums[slow] # important!
            
            
            if slow == k:
                return nums[slow]
            elif slow > k:
                return help(start, slow - 1, k)
            elif slow < k:
                return help(slow + 1, end, k) # must start with slow + 1 to avoid infinite loop when where are equal elements
            
        return help(0, len(nums) - 1, len(nums) - k)

```

### 88. Merge Sorted Array

```python
Given two sorted integer arrays nums1 and nums2, merge nums2 into nums1 as one sorted array.

Note:

The number of elements initialized in nums1 and nums2 are m and n respectively.
You may assume that nums1 has enough space (size that is greater or equal to m + n) to hold additional elements from nums2.

class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        
        slow = m + n - 1
        i, j = m - 1, n - 1
        while i >= 0 and j >= 0:
            
            if nums1[i] < nums2[j]:
                nums1[slow] = nums2[j]
                slow -= 1
                j -= 1
            else:
                nums1[slow] = nums1[i]
                slow -= 1
                i -= 1
                
        while j >= 0:
            nums1[slow] = nums2[j]
            slow -= 1
            j -= 1
```

### Union Two-Sorted Arrays (remove duplicates)

```python
# arr1 = [1, 3, 3, 5, 7]
# arr2 = [1, 3, 4, 5, 7, 8, 8, 8]
# Union of both the arrays
# res = [1, 3, 4, 5, 7, 8]

def union(arr1, arr2):
    
    p1 = 0
    p2 = 0
    res = []
    while p1 < len(arr1) and p2 < len(arr2):
        
        if arr1[p1] <= arr2[p2]:
            # only push the unique one
            if not res or arr1[p1] != res[-1]:
                res.append(arr1[p1])
            p1 += 1
        elif arr1[p1] > arr2[p2]:
            if not res or arr2[p2] != res[-1]:
                res.append(arr2[p2])
            p2 += 1
            
    while p1 < len(arr1):
        if not res or arr1[p1] != res[-1]:
            res.append(arr1[p1])
        p1 += 1
            
    while p2 < len(arr2):
        if not res or arr2[p2] != res[-1]:
            res.append(arr2[p2])
        p2 += 1
    return res
```

```python
class Solution:
    def arraysIntersection(self, arr1: List[int], arr2: List[int], arr3: List[int]) -> List[int]:
        
        
        result = []
        
        i = j = k = 0
        
        while i < len(arr1) and j < len(arr2) and k < len(arr3):
            minVal = min(arr1[i], arr2[j], arr3[k])
            if arr1[i] == arr2[j] == arr3[k]:
                if not result or arr1[i] != result[-1]:
                    result.append(arr1[i])
                i += 1
                j += 1
                k += 1
            elif arr1[i] == minVal:
                i += 1
            elif arr2[j] == minVal:
                j += 1
            elif arr3[k] == minVal:
                k += 1
                
        return result
                            
                                 
```

### 244. Shortest Word Distance II

Design a class which receives a list of words in the constructor, and implements a method that takes two words word1 and word2 and return the shortest distance between these two words in the list. Your method will be called repeatedly many times with different parameters. 

```python
class WordDistance:

    def __init__(self, words: List[str]):
        self.pos = collections.defaultdict(list)
        for i, w in enumerate(words):
            self.pos[w].append(i)
        
        self.memo = {}

    def shortest(self, word1: str, word2: str) -> int:
        if word1 > word2:
            word1, word2 = word2, word1
        if (word1, word2) in self.memo:
            return self.memo[(word1, word2)]
        
        list1 = self.pos[word1]
        list2 = self.pos[word2]
        p1, p2 = 0, 0
        minDist = float('inf')
        while p1 < len(list1) and p2 < len(list2):
            minDist = min(minDist, abs(list1[p1] - list2[p2]))
            if list1[p1] < list2[p2]:
                p1 += 1
            else:
                p2 += 1
        self.memo[(word1, word2)] = minDist
        return minDist
```

### 245. Shortest Word Distance III

Given a list of words and two words word1 and word2, return the shortest distance between these two words in the list.

word1 and word2 may be the same and they represent two individual words in the list.

```python
class Solution:
    def shortestWordDistance(self, words: List[str], word1: str, word2: str) -> int:
        
        p1, p2 = -1, -1
        minDist = len(words)
        
        for i in range(len(words)):
            if words[i] == word1:
                # if equal, then p2 records the previous
                if word1 == word2:
                    p2 = p1
                p1 = i
            elif words[i] == word2:
                p2 = i
            
            if p1 >= 0 and p2 >= 0:
                minDist = min(minDist, abs(p1 - p2))
        return minDist
```

### 209. Minimum Size Subarray Sum at least K (only positive integers)

```python
209. Minimum Size Subarray Sum
Given an array of n positive integers and a positive integer s, find the minimal length of a contiguous subarray of which the sum  s. If there isn't one, return 0 instead.

Example: 

Input: s = 7, nums = [2,3,1,2,4,3]
Output: 2
Explanation: the subarray [4,3] has the minimal length under the problem constraint.


# the condition of positive interger are critical
class Solution:
    def minSubArrayLen(self, s: int, nums: List[int]) -> int:
        
        minLen = float('inf')
        
        sum = 0
        slow = 0
        
        for fast in range(len(nums)):
            
            sum += nums[fast]
            # when sum >= s, no need to move fast further since we are seeking the minimum array size
            # we can increase slow like we are doing a double iteration
            # say we increase slow from 0 to 1 but still sum >= s, do we need to decrease fast?
            # we do not! because sum(nums[0:fast]) >= s and  sum(nums[0:fast - 1]) < s, 
            # sum(nums[1:fast]) could be greater or smaller than s. BUT, sum(nums[1:fast - 1]) will definitely smaller than s
            # after we increase slow to sum(nums[slow + 1:fast]) < s, we can continue to crease fast
            while sum >= s:
                minLen = min(minLen, fast - slow + 1)
                sum -= nums[slow]
                slow += 1
        
        return minLen if minLen != float('inf') else 0
```

### 792. Number of Matching Subsequences
Given string S and a dictionary of words words, find the number of words[i] that is a subsequence of S.

```python
# A brute force approach
class Solution:
    def numMatchingSubseq(self, S: str, words: List[str]) -> int:
        
        p = [0] * len(words)
        for ps in range(len(S)):
            for i in range(len(words)):
                if p[i] < len(words[i]) and words[i][p[i]] == S[ps]:
                    p[i] += 1
        return sum( [p[i] == len(words[i]) for i in range(len(words))])
    
# An highly optimized iterator approach with the idea of graph traversal
class Solution:
    def numMatchingSubseq(self, S: str, words: List[str]) -> int:
        
        p = [0] * len(words)
        
        nextIdx = collections.defaultdict(list)
        
        for i, w in enumerate(words):
            nextIdx[w[0]].append(i)
        
        answer = 0
        for ps in range(len(S)):
            c = S[ps]
            tempList = nextIdx[c]
            nextIdx[c] = []
            for i in tempList:
                if p[i] == len(words[i]) - 1:
                    answer += 1
                else:
                    p[i] += 1
                    nextChar = words[i][p[i]]
                    nextIdx[nextChar].append(i)
        return answer
        
```

```python
3Sum Closest

Given an array nums of n integers and an integer target, find three integers in nums such that the sum is closest to target. Return the sum of the three integers. You may assume that each input would have exactly one solution.

Example:

Given array nums = [-1, 2, 1, -4], and target = 1.

The sum that is closest to the target is 2. (-1 + 2 + 1 = 2).


class Solution:
    def threeSumClosest(self, nums: List[int], target: int) -> int:
        
        
        nums.sort()
        
        minSum = float('inf')
        
        for i in range(len(nums)):
            
            newTarget = target - nums[i]
            
            left = i + 1
            right = len(nums) - 1
            
            while left < right:
                
                sum = nums[left] + nums[right]
                
                if abs(sum - newTarget) < abs(minSum - target):
                    minSum = sum + nums[i]
                
                if sum < newTarget:
                    left += 1
                elif sum > newTarget:
                    right -= 1
                elif sum == newTarget:
                    return minSum
        
        return minSum
```

### 986. Interval List Intersections

```python
986. Interval List Intersections

Given two lists of closed intervals, each list of intervals is pairwise disjoint and in sorted order.

Return the intersection of these two interval lists.

(Formally, a closed interval [a, b] (with a <= b) denotes the set of real numbers x with a <= x <= b. 
 The intersection of two closed intervals is a set of real numbers that is either empty, or can be represented
 as a closed interval.  For example, the intersection of [1, 3] and [2, 4] is [2, 3].)

class Solution:
    def intervalIntersection(self, A: List[List[int]], B: List[List[int]]) -> List[List[int]]:
        
        i = 0
        j = 0
        
        res = []
        
        while i < len(A) and j < len(B):
            
            lo = max(A[i][0], B[j][0])
            hi = min(A[i][1], B[j][1])

            if lo <= hi:
                res.append([lo, hi])
                
        
            if A[i][1] <= B[j][1]:
                i += 1
            else:
                j += 1
                
        return res
                
                
                
```

```python

class Solution:
    def numSubarrayProductLessThanK(self, nums: List[int], k: int) -> int:
        
        if k <= 1:
            return 0
        
        count = 0
        prod = 1
        slow = 0
        for fast in range(len(nums)):
            
            prod *= nums[fast]
            
            while prod >= k:
                prod /= nums[slow]
                slow += 1
                
            count += fast - slow + 1
            
        return count
```

### 930. Number of Binary Subarrays With Sum K
In an array A of 0s and 1s, how many non-empty subarrays have sum S?

```python
# note that this method only applies to non-negative arrays.
class Solution:
    def numSubarraysWithSum(self, A: List[int], S: int) -> int:
        
        def atMost(k):
            # critical
            if k < 0: return 0
            slow = 0
            preSum = 0
            answer = 0
            for fast in range(len(A)):
                preSum += A[fast]
                # if preSum > k
                while preSum > k:
                    preSum -= A[slow]
                    slow += 1
                # because nonnegativeness, if we find a subarray[slow: fast] whose sum <= K, then 
                # there are totally fast - slow + 1 subarrays
                answer += fast - slow + 1
            return answer
        return atMost(S) - atMost(S - 1)
```

### 1438. Longest Continuous Subarray With Absolute Diff Less Than or Equal to Limit
Given an array of integers nums and an integer limit, return the size of the longest continuous subarray such that the absolute difference between any two elements is less than or equal to limit.

In case there is no subarray satisfying the given condition return 0.

```python

# use monotone queues to maintain running max and min
class Solution:
    def longestSubarray(self, nums: List[int], limit: int) -> int:
        
        
        slow = 0
        answer = 1
        queue1 = collections.deque([nums[0]])
        queue2 = collections.deque([nums[0]])
        
        for fast in range(1, len(nums)):
         
            while queue1 and nums[fast] > queue1[-1]:
                queue1.pop()
            queue1.append(nums[fast])
            
            while queue2 and nums[fast] < queue2[-1]:
                queue2.pop()
            queue2.append(nums[fast])
            
            while queue1[0] - queue2[0] > limit:
                
                if nums[slow] == queue1[0]:
                    queue1.popleft()

                if nums[slow] == queue2[0]:
                    queue2.popleft()
                    
                
                slow += 1
                
            answer = max(answer, fast - slow + 1)
            
        return answer
            
```

### 862. Shortest Subarray with Sum at Least K (contains both positive and negative values)
Return the length of the shortest, non-empty, contiguous subarray of A with sum at least K.

If there is no non-empty subarray with sum at least K, return -1.

```python
# use a monotine queue to maintain an increasing presum sequence
class Solution:
    def shortestSubarray(self, A: List[int], K: int) -> int:
        
        queue = collections.deque([(0, -1)])
        answer = float('inf')
        preSum = 0
        for i in range(len(A)):
            
            preSum += A[i]    
            while queue and preSum <= queue[-1][0]:
                queue.pop()
            queue.append((preSum, i))
            
            while preSum - queue[0][0] >= K:
                answer = min(answer, i - queue[0][1])
                queue.popleft()
        return answer if answer != float('inf') else -1
```

### 795. Number of Subarrays with Bounded Maximum
We are given an array A of positive integers, and two positive integers L and R (L <= R).

Return the number of (contiguous, non-empty) subarrays such that the value of the maximum array element in that subarray is at least L and at most R.

```python
# use a monotine queue to maintain an decraseing sequence to enable retrive sliding max value
class Solution:
    def numSubarrayBoundedMax(self, A: List[int], L: int, R: int) -> int:
        if not A:
            return 0
        
        def atMost(R):
            res = 0
            queue = collections.deque()
            slow = 0
            for fast in range(len(A)):
                while queue and A[fast] >= A[queue[-1]]:
                    queue.pop()
                queue.append(fast)

                while queue and A[queue[0]] > R:
                    if slow == queue[0]:
                        queue.popleft()
                    slow += 1

                res += fast - slow + 1
            return res
        return atMost(R) - atMost(L - 1)
                

# a much simpler solution
class Solution:
    def numSubarrayBoundedMax(self, A: List[int], L: int, R: int) -> int:
        if not A:
            return 0
        
        def atMost(R):
            answer = 0
            
            count = 0
            for fast in range(len(A)):
                
                if A[fast] > R:
                    # reset to 0; restart counting
                    count = 0
                else:
                    count += 1
                    
                answer += count
            return answer
        return atMost(R) - atMost(L - 1)
            
                
```

```python
19. Remove Nth Node From End of List

Given a linked list, remove the n-th node from the end of list and return its head.

class Solution:
    def removeNthFromEnd(self, head: ListNode, n: int) -> ListNode:
        
        slow = fast = head
        
        dummyHead = ListNode(-1)
        dummyHead.next = head
        prev = dummyHead
        
        for _ in range(n):
            fast = fast.next
            
        while fast:
            prev = slow
            slow = slow.next
            fast = fast.next
            
        prev.next = slow.next
        
        return dummyHead.next
```

```python
142. Linked List Cycle II
Given a linked list, return the node where the cycle begins. If there is no cycle, return null.

To represent a cycle in the given linked list, we use an integer pos which represents the position (0-indexed) in the linked list where tail connects to. If pos is -1, then there is no cycle in the linked list.

Note: Do not modify the linked list.

class Solution:
    def detectCycle(self, head: ListNode) -> ListNode:
        
        if not head:
            return head
        
        
        slow = head
        fast = head
        intersection = None
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                intersection = slow
                break
                
        if not intersection:
            return None
        
        p1 = head
        p2 = intersection
        
        while p1 != p2:
            p1 = p1.next
            p2 = p2.next
            
        return p1
        
```

```python
287. Find the Duplicate Number
Given an array nums containing n + 1 integers where each integer is between 1 and n (inclusive), prove that at least one duplicate number must exist. Assume that there is only one duplicate number, find the duplicate one.




class Solution:
    def findDuplicate(self, nums: List[int]) -> int:
        
        slow = nums[0]
        fast = nums[0]
        
        while True:
            slow = nums[slow]
            fast = nums[nums[fast]]
            if slow == fast:
                break
            
        p1 = nums[0]
        p2 = slow
        
        while p1 != p2:
            p1 = nums[p1]
            p2 = nums[p2]
            
        return p1
```

### 1229. Meeting Scheduler

```python
Given the availability time slots arrays slots1 and slots2 of two people and a meeting duration duration, return the earliest time slot that works for both of them and is of duration duration.

If there is no common time slot that satisfies the requirements, return an empty array.

The format of a time slot is an array of two elements [start, end] representing an inclusive time range from start to end.  

It is guaranteed that no two availability slots of the same person intersect with each other. That is, for any two time slots [start1, end1] and [start2, end2] of the same person, either start1 > end2 or start2 > end1.

 class Solution:
    def minAvailableDuration(self, slots1: List[List[int]], slots2: List[List[int]], duration: int) -> List[int]:
        
        slots1.sort(key=lambda x: (x[0], x[1]))
        slots2.sort(key=lambda x: (x[0], x[1]))
        
        first, second = 0, 0
        
        while first < len(slots1) and second < len(slots2):
            
            p1 = slots1[first]
            p2 = slots2[second]
            overLap = min(p1[1], p2[1]) - max(p1[0], p2[0])
            
            if overLap >= duration:
                return [max(p1[0], p2[0]), max(p1[0], p2[0]) + duration]
            
            if p1[1] <= p2[1]:
                first += 1
            else:
                second += 1
                
                
        return []
                
```

### 986. Interval List Intersections

```python
Given two lists of closed intervals, each list of intervals is pairwise disjoint and in sorted order.

Return the intersection of these two interval lists.

(Formally, a closed interval [a, b] (with a <= b) denotes the set of real numbers x with a <= x <= b.  The intersection of two closed intervals is a set of real numbers that is either empty, or can be represented as a closed interval.  For example, the intersection of [1, 3] and [2, 4] is [2, 3].)

class Solution:
    def intervalIntersection(self, A: List[List[int]], B: List[List[int]]) -> List[List[int]]:
        
        
        first, second = 0, 0
        res = []
        while first < len(A) and second < len(B):
            
            lo = min(A[first][1], B[second][1])
            hi = max(A[first][0], B[second][0])
            
            if lo >= hi:
                res.append([hi, lo])
                
            if A[first][1] < B[second][1]:
                first += 1
            else:
                second += 1
                
        return res
                
```
