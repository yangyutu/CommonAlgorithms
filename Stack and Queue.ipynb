```python
155. Min Stack
Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

push(x) -- Push element x onto stack.
pop() -- Removes the element on top of the stack.
top() -- Get the top element.
getMin() -- Retrieve the minimum element in the stack.


# one stack solution
class MinStack:

    def __init__(self):
        """
        initialize your data structure here.
        """
        self.stack = []
        

    def push(self, x: int) -> None:
        currMin = self.stack[-1][1] if self.stack else float('inf')
        self.stack.append((x, min(currMin, x)))
        

    def pop(self) -> None:
        if self.stack:
            return self.stack.pop()[0]
        else:
            return None
        

    def top(self) -> int:
        if self.stack:
            return self.stack[-1][0]
        else:
            return None
        

    def getMin(self) -> int:
        if self.stack:
            return self.stack[-1][1]
        else:
            return None
        
# two stack solution
class MinStack:

    def __init__(self):
        """
        initialize your data structure here.
        """
        self.stack = []
        self.minStack = []

    def push(self, x: int) -> None:
        
        self.stack.append(x)
        if not self.minStack or x <= self.minStack[-1]:
            self.minStack.append(x)

    def pop(self) -> None:
        
        val = self.stack.pop()
        if val == self.minStack[-1]:
            self.minStack.pop()

        

    def top(self) -> int:
        
        
        return self.stack[-1]

    def getMin(self) -> int:
        
        return self.minStack[-1]


# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(x)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```

### 716. Max Stack

```python
716. Max Stack
Design a max stack that supports push, pop, top, peekMax and popMax.

push(x) -- Push element x onto stack.
pop() -- Remove the element on top of the stack and return it.
top() -- Get the element on the top.
peekMax() -- Retrieve the maximum element in the stack.
popMax() -- Retrieve the maximum element in the stack, and remove it. If you find more than one maximum elements, 
only remove the top-most one.



The last four operations won't be called when stack is empty.

class MaxStack:

    def __init__(self):
        """
        initialize your data structure here.
        """
        self.stackData = []
        self.maxData = []
        

    def push(self, x: int) -> None:
        self.stackData.append(x)
        if not self.maxData or x >= self.maxData[-1]:
            self.maxData.append(x)

        

    def pop(self) -> int:
        res = self.stackData.pop()
        if res == self.maxData[-1]:
            self.maxData.pop()
            
        return res
        

    def top(self) -> int:
        return self.stackData[-1]

    def peekMax(self) -> int:
        return self.maxData[-1]

    ## this is most important/chanllging part
    def popMax(self) -> int:
        temp = []
        while self.stackData[-1] != self.maxData[-1]:
            temp.append(self.stackData.pop())
        res = self.stackData.pop()
        self.maxData.pop()
        for e in temp[::-1]:
            self.push(e)
        return res
```

### 622. Design Circular Queue
Design your implementation of the circular queue. The circular queue is a linear data structure in which the operations are performed based on FIFO (First In First Out) principle and the last position is connected back to the first position to make a circle. It is also called "Ring Buffer".

One of the benefits of the circular queue is that we can make use of the spaces in front of the queue. In a normal queue, once the queue becomes full, we cannot insert the next element even if there is a space in front of the queue. But using the circular queue, we can use the space to store new values.


```python

Your implementation should support following operations:

MyCircularQueue(k): Constructor, set the size of the queue to be k.
Front: Get the front item from the queue. If the queue is empty, return -1.
Rear: Get the last item from the queue. If the queue is empty, return -1.
enQueue(value): Insert an element into the circular queue. Return true if the operation is successful.
deQueue(): Delete an element from the circular queue. Return true if the operation is successful.
isEmpty(): Checks whether the circular queue is empty or not.
isFull(): Checks whether the circular queue is full or not.

class MyCircularQueue:

    def __init__(self, k: int):
        """
        Initialize your data structure here. Set the size of the queue to be k.
        """
        self.capacity = k
        self.data = [0] * k
        self.front = 0
        self.size = 0


    def enQueue(self, value: int) -> bool:
        """
        Insert an element into the circular queue. Return true if the operation is successful.
        """
        if not self.isFull():
            self.data[(self.front + self.size)%self.capacity] = value
            self.size += 1
            return True
        return False

    def deQueue(self) -> bool:
        """
        Delete an element from the circular queue. Return true if the operation is successful.
        """
        if self.size > 0:
            self.front += 1
            self.size -= 1
            return True
        return False

    def Front(self) -> int:
        """
        Get the front item from the queue.
        """
        if self.size > 0:
            return self.data[self.front]
        else:
            return -1

    def Rear(self) -> int:
        """
        Get the last item from the queue.
        """
        if self.size > 0:
            return self.data[(self.front + self.size - 1)%self.capacity]
        else:
            return -1

    def isEmpty(self) -> bool:
        """
        Checks whether the circular queue is empty or not.
        """
        return self.size == 0

    def isFull(self) -> bool:
        """
        Checks whether the circular queue is full or not.
        """
        return self.size == self.capacity


# Your MyCircularQueue object will be instantiated and called as such:
# obj = MyCircularQueue(k)
# param_1 = obj.enQueue(value)
# param_2 = obj.deQueue()
# param_3 = obj.Front()
# param_4 = obj.Rear()
# param_5 = obj.isEmpty()
# param_6 = obj.isFull()


# singly linked list implementation

class ListNode:
    def __init__(self, value):
        self.val = value
        self.next = None


class MyCircularQueue:

    def __init__(self, k: int):
        """
        Initialize your data structure here. Set the size of the queue to be k.
        """
        self.capacity = k
        self.size = 0
        self.head, self.tail = None, None

    def enQueue(self, value: int) -> bool:
        """
        Insert an element into the circular queue. Return true if the operation is successful.
        """
        if not self.isFull():
            newNode = ListNode(value)
            if self.size == 0:
                self.tail = newNode
                self.head = newNode
            else:
                self.tail.next = newNode
                self.tail = newNode
            self.size += 1
            return True
        return False
        
        

    def deQueue(self) -> bool:
        """
        Delete an element from the circular queue. Return true if the operation is successful.
        """
        if self.size > 0:
            self.head = self.head.next
            self.size -= 1
            if self.size == 0:
                self.tail = None
            return True
        return False
            
        
        

    def Front(self) -> int:
        """
        Get the front item from the queue.
        """
        if self.size > 0:
            return self.head.val
        return -1

    def Rear(self) -> int:
        """
        Get the last item from the queue.
        """
        if self.size > 0:
            return self.tail.val
        return -1

    def isEmpty(self) -> bool:
        """
        Checks whether the circular queue is empty or not.
        """
        return self.size == 0

    def isFull(self) -> bool:
        """
        Checks whether the circular queue is full or not.
        """
        return self.size == self.capacity


# Your MyCircularQueue object will be instantiated and called as such:
# obj = MyCircularQueue(k)
# param_1 = obj.enQueue(value)
# param_2 = obj.deQueue()
# param_3 = obj.Front()
# param_4 = obj.Rear()
# param_5 = obj.isEmpty()
# param_6 = obj.isFull()
```

### 150. Evaluate Reverse Polish Notation
Evaluate the value of an arithmetic expression in Reverse Polish Notation.

Valid operators are +, -, *, /. Each operand may be an integer or another expression.

Input: ["2", "1", "+", "3", "*"]
Output: 9
Explanation: ((2 + 1) * 3) = 9


```python
class Solution:
    def evalRPN(self, tokens: List[str]) -> int:
        
        stack = []
        signs = ['+', '-', '*', '/']
        for e in tokens:
            
            if e not in signs:
                stack.append(e)
            else:
                op2 = int(stack.pop())
                op1 = int(stack.pop())
                if e == '+':
                    stack.append(op1 + op2)
                elif e == '-':
                    stack.append(op1 - op2)
                elif e == '*':
                    stack.append(op1 * op2)
                else:
                    stack.append(int(op1 / op2))
                    
        return stack[0]
                    
                        
```

### 1249. Minimum Remove to Make Valid Parentheses
Given a string s of '(' , ')' and lowercase English characters. 

Your task is to remove the minimum number of parentheses ( '(' or ')', in any positions ) so that the resulting parentheses string is valid and return any valid string.

Formally, a parentheses string is valid if and only if:

It is the empty string, contains only lowercase characters, or
It can be written as AB (A concatenated with B), where A and B are valid strings, or
It can be written as (A), where A is a valid string.


```python
# two pass solution, using extra array to store idx of parenthesis that causes imbalance
class Solution:
    def minRemoveToMakeValid(self, s: str) -> str:
        
        
        stack = []
        toRemove = set()
        for i, c in enumerate(s):
            
            if c == '(':
                stack.append(i)
            elif c == ')':
                if stack:
                    stack.pop()
                else:
                    toRemove.add(i)
                    
        toRemove.update(stack)
        
        ans = []
        
        for i, c in enumerate(s):
            if i not in toRemove:
                ans.append(c)
                
        return ''.join(ans)
            
# two pass solution without stack
class Solution:
    def minRemoveToMakeValid(self, s: str) -> str:
        
        res = list(s)
        open = 0
        for i, c in enumerate(s):
            
            if c == '(':
                open += 1
            elif c == ')':
                if open == 0:
                    res[i] = ""
                elif open > 0:
                    open -= 1
        
        # must from back to front
        for i in range(len(res) - 1, -1, -1):
            
            if open == 0:
                break
            else:
                if res[i] == '(':
                    res[i] = ""
                    open -= 1
                    
        return ''.join(res)      
```

### 856. Score of Parentheses

Given a balanced parentheses string S, compute the score of the string based on the following rule:

() has score 1
AB has score A + B, where A and B are balanced parentheses strings.
(A) has score 2 * A, where A is a balanced parentheses string.

```python
# recursive solution
class Solution:
    def scoreOfParentheses(self, S: str) -> int:
        
        def help():  
            total = 0
            while self.idx < len(S):
                c = S[self.idx]
                
                if c == '(':
                    self.idx += 1
                    score = help()
                    if score == 0:
                        total += 1
                    else:
                        total += 2 * score
                elif c == ')':
                    return total
                
                self.idx += 1
            return total
                
        self.idx = 0
        total = help()
        
        return total

# stack solution
class Solution:
    def scoreOfParentheses(self, S: str) -> int:
        
        
        total = 0
        stack = []
        
        for c in S:
             
            if c == '(':
                stack.append(total)
                total = 0
            elif c == ')':
                if total == 0:
                    total = 1 + stack.pop()
                else:
                    total = 2 * total + stack.pop()
        return total        
```

### 1190. Reverse Substrings Between Each Pair of Parentheses
You are given a string s that consists of lower case English letters and brackets. 

Reverse the strings in each pair of matching parentheses, starting from the innermost one.

Your result should not contain any brackets.

```python
class Solution:
    def reverseParentheses(self, s: str) -> str:
        
        stack = []
        cur = ''
        for c in s:
            if c == '(':
                stack.append(cur)
                cur = ''
            elif c == ')':
                cur = stack.pop() + cur[::-1]
            else:
                cur = cur + c
        return cur
```

### 224. Basic Calculator
Implement a basic calculator to evaluate a simple expression string.

The expression string may contain open ( and closing parentheses ), the plus + or minus sign -, 
non-negative integers and empty spaces .

```python
# stack solution
class Solution:
    def calculate(self, s: str) -> int:
        
        stack = []  
        operand = 0
        sign = 1
        res = 0
        
        for c in s:
            if c.isdigit():
                operand = (operand * 10 + int(c))   
            elif c == '+':  
                res += sign * operand
                sign = 1
                operand = 0
            elif c == '-':
                res += sign * operand
                sign = -1
                operand = 0
            elif c == '(':
                stack.append(res)
                stack.append(sign)
                sign = 1
                res = 0
            elif c == ')':
                res += sign * operand
                # first pop is sign
                res = stack.pop() * res
                # second pop is previous value
                res = stack.pop() + res
                operand = 0
                sign = 1 # this sign could be removed
        res += sign * operand
        
        return res
    
class Solution:
    def calculate(self, s: str) -> int:
        
        # global index to control the process
        self.idx = 0
        
        def help(s):
            operand = 0
            sign = 1
            res = 0

            while self.idx < len(s):
                c = s[self.idx]
                if c.isdigit():
                    operand = (operand * 10 + int(c))
                elif c == '+':  
                    res += sign * operand
                    sign = 1
                    operand = 0
                elif c == '-':
                    res += sign * operand
                    sign = -1
                    operand = 0
                elif c == '(':
                    self.idx += 1
                    operand = help(s) 
                elif c == ')':
                    res += sign * operand
                    return res
                self.idx += 1 
            res += sign * operand
            return res
        
        return help(s)
```

### 227. Basic Calculator II
Implement a basic calculator to evaluate a simple expression string.

```python
239. Sliding Window Maximum

Given an array nums, there is a sliding window of size k which is moving from the very left of the array to the very right. You can only see the k numbers in the window.
Each time the sliding window moves right by one position. Return the max sliding window.

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
               
        queue = collections.deque()
        res = []
        for i in range(len(nums)):
            # remove out of window one
            if queue and queue[0][1] == i - k:
                queue.popleft()
            
            # remove smaller one since they can never be the result
            while queue and queue[-1][0] <= nums[i]:
                queue.pop()
                
            queue.append((nums[i], i))
                        
            if i >= k - 1:
                res.append(queue[0][0])
                
                
        return res
        
                
```

### 1019. Next Greater Node In Linked List

```python
We are given a linked list with head as the first node.  Let's number the nodes in the list: node_1, node_2, node_3, ... etc.

Each node may have a next larger value: for node_i, next_larger(node_i) is the node_j.val such that j > i, node_j.val > node_i.val, and j is the smallest possible choice.  If such a j does not exist, the next larger value is 0.

Return an array of integers answer, where answer[i] = next_larger(node_{i+1}).

Note that in the example inputs (not outputs) below, arrays such as [2,1,5] represent the serialization of a linked list with a head node value of 2, second node value of 1, and third node value of 5.

class Solution:
    def nextLargerNodes(self, head: ListNode) -> List[int]:
        
        res = []
        stack = []
        idx = 0
        # maintain a decreasing stack
        while head:
            while stack and head.val > stack[-1][1]:
                # kick, since larger elements found
                res[stack[-1][0]] = head.val
                stack.pop()
            
            stack.append((idx, head.val))
            res.append(0)
            head = head.next
            idx += 1
        
        return res
        
```

```python
English
Given an integer array with no duplicates. A max tree building on this array is defined as follow:

The root is the maximum number in the array
The left subtree and right subtree are the max trees of the subarray divided by the root number.
Construct the max tree by the given array.

Input[2, 5, 6, 0, 3, 1]
Output{6,5,3,2,#,0,1}
Explanation
the max tree constructed by this array is:
    6
   / \
  5   3
 /   / \
2   0   1

# A simple approach is to look for the maxmum element as the root, and recusively work on left and right
# Using one-pass monotonic stack, for each element, we look for the largest smaller on the left and right
# largest smaller is the opposite of next greater on the left and right.
 
class Solution:
    """
    @param A: Given an integer array with no duplicates.
    @return: The root of max tree.
    """
    def maxTree(self, A):
        # write your code here

        A.append(float('inf'))
        
        nodes = [TreeNode(e) for e in A]
        
        numNodes = len(nodes)
        
        stack = []
        
        for i in range(len(A)):
            
            while stack and A[i] > A[stack[-1]]:
                
                # next bigger on the right of idx is [i]
                idx = stack.pop()
                
                # if we have next bigger on the left and the left is smaller then right
                if not stack:
                    nodes[i].left = nodes[idx]
                elif stack and A[stack[-1]] < A[i]:
                    nodes[stack[-1]].right = nodes[idx]
                elif stack and A[stack[-1]] > A[i]:
                    nodes[i].left = nodes[idx]
                    

            stack.append(i)
        
    return nodes[-1].left
        
# the optimal solution       
class Solution:
    def constructMaximumBinaryTree(self, A: List[int]) -> TreeNode:
    
        if not A:
            return None
        
        stack = []
        
        
        for i in range(len(A)):
            
            curr = TreeNode(A[i])
            # maintain the decreasing stack
            # curr.left finally connect to the one that is the largest smaller element on the left
            while stack and A[i] > stack[-1].val:
                curr.left = stack.pop()
            
            # stack[-1].right connects to the largest smaller element on the right
            if stack:
                stack[-1].right = curr
                
            stack.append(curr)
            # the largest element is on the front
        return stack[0]
```

### 84. Largest Rectangle in Histogram

```python
Given n non-negative integers representing the histogram's bar height where the 
width of each bar is 1, find the area of largest rectangle in the histogram.


# one pass to get smaller elements on both sides
class Solution:
    def largestRectangleArea(self, heights: List[int]) -> int:
        
        
        maxVal = 0
        
        stack = []
        
        candidates = []
        
        for i in range(len(heights)):
            
            while stack and heights[stack[-1]] > heights[i]:
                
                idx = stack.pop()
                right = i - 1
                if stack:
                    left = stack[-1] + 1
                else:
                    left = 0
                    
                candidates.append((idx, left, right))
                
            stack.append(i)
        
        
        while stack:
            idx = stack.pop()
            right = len(heights) - 1
            if stack:
                left = stack[-1] + 1
            else:
                left = 0        
            candidates.append((idx, left, right))
        
        for idx, left, right in candidates:
            maxVal = max(maxVal, heights[idx] * (right - left + 1))
            
        
        return maxVal






# one-pass O(1) optimized solution
class Solution:
    def largestRectangleArea(self, heights: List[int]) -> int:
        
        
        maxVal = 0
        
        stack = []
        heights.append(-1)
        for i in range(len(heights)):
            
            while stack and heights[stack[-1]] > heights[i]:
                
                idx = stack.pop()
                right = i - 1
                if stack:
                    left = stack[-1] + 1
                else:
                    left = 0
                maxVal = max(maxVal, heights[idx] * (right - left + 1))
            stack.append(i)
        
        return maxVal
```

### 1130. Minimum Cost Tree From Leaf Values

```python
Given an array arr of positive integers, consider all binary trees such that:

Each node has either 0 or 2 children;
The values of arr correspond to the values of each leaf in an in-order traversal of the tree.  (Recall that a node is a leaf if and only if it has 0 children.)
The value of each non-leaf node is equal to the product of the largest leaf value in its left and right subtree respectively.
Among all possible binary trees considered, return the smallest possible sum of the values of each non-leaf node.  It is guaranteed this sum fits into a 32-bit integer.

 

Example 1:

Input: arr = [6,2,4]
Output: 32
Explanation:
There are two possible trees.  The first has non-leaf node sum 36, and the second has non-leaf node sum 32.

    24            24
   /  \          /  \
  12   4        6    8
 /  \               / \
6    2             2   4
    
class Solution:
    def mctFromLeafValues(self, arr: List[int]) -> int:
        
        nextGreaterLeft = []
        nextGreaterRight = []
        
        res = 0
        
        stack = []
        
        for i in range(len(arr)):
            
            while stack and stack[-1] <= arr[i]:
                
                right = arr[i]
                element = stack.pop()
                
                if stack:
                    left = stack[-1]
                else:
                    left = float('inf')
                
                res += element * min(left, right)
            
            stack.append(arr[i])
            
        while len(stack) > 1:
            
            element = stack.pop()
            left = stack[-1]
            res += element * left
            
        return res
```

###  907. Sum of Subarray Minimums

Since the answer may be large, return the answer modulo 10^9 + 7.

 

Example 1:

Input: [3,1,2,4]
Output: 17
Explanation: Subarrays are [3], [1], [2], [4], [3,1], [1,2], [2,4], [3,1,2], [1,2,4], [3,1,2,4]. 
Minimums are 3, 1, 2, 4, 1, 1, 2, 1, 1, 1.  Sum is 17.

### 670. Maximum Swap

Given a non-negative integer, you could swap two digits at most once to get the maximum valued number. Return the maximum valued number you could get.

```python
class Solution:
    def maximumSwap(self, num: int) -> int:
        string = list(str(num))
        numList = [int(e) for e in string]
        
        queue = collections.deque()
        
        for i in range(1, len(numList)):
            while queue and numList[i]>= numList[queue[-1]]:
                queue.pop()
            queue.append(i)
            
        for i in range(len(numList)):
            if queue and numList[i] < numList[queue[0]]:
                idx = queue[0]
                numList[i], numList[idx] = numList[idx], numList[i]
                return int(''.join([str(e) for e in numList]))
            if queue and queue[0] == i + 1:
                queue.popleft()
        return num
```
